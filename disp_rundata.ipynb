```python
import pandas as pd
import pickle
```

```python
data = pd.DataFrame()
data = data.append(pd.read_csv("data/0_results.csv"))
data = data.append(pd.read_csv("data/3_results.csv"))
```

```python
data = data.drop("Unnamed: 0",axis=1)
data.to_csv("data/combined.csv")
```

```python
data
```

```python
criterea = ['HiddenLayers', 'LearningRate', 'Optimizer', 'NumFilters', 'Activation', 'KernelSize', 'Momentum']
```

```python
grouped_df = data.groupby(criterea)

groups = []
for key,item in grouped_df:
    groups += [grouped_df.get_group(key)]
    
print(len(groups))
```

```python
import plotly.graph_objects as go


fig = go.Figure()

x_l = "Epochs"
y_l = "Acc"

for g in groups:
    x = g[x_l]
    y = g[y_l]
    label = "".join([str(g[i].unique()[0]) + ";" for i in criterea])
    fig.add_trace(go.Scatter(x=x, y=y,
                        mode='lines',
                        name=label))
print(criterea)

fig.update_layout(
    title=f"{y_l} over {x_l}",
    xaxis_title=x_l,
    yaxis_title=y_l,
    legend_title="Model Parameters",
)

fig.show()
```

```python
vary = "HiddenLayers"

v_idx = criterea.index(vary)
sel_crit = criterea[:v_idx] + criterea[v_idx + 1:]
grouped_df = data.groupby(sel_crit)

sel_groups = []
for key,item in grouped_df:
    sel_groups += [grouped_df.get_group(key)]
        
fig = go.Figure()

x_l = vary
y_l = "Acc"

for g in sel_groups:
    x = list(g[x_l].unique())
    y = [max(g[g[x_l] == i][y_l]) for i in x]
    
    x, y = [list(i) for i in list(zip(*sorted(zip(x, y), key=lambda x: x[0])))]
    
    label = "".join([str(g[i].unique()[0]) + ";" for i in sel_crit])
    fig.add_trace(go.Scatter(x=x, y=y,
                        mode='lines+markers',
                        name=label))
print(criterea)

fig.update_layout(
    title=f"{y_l} over {x_l}",
    xaxis_title=x_l,
    yaxis_title=y_l,
    legend_title="Model Parameters",
)

fig.show()
```
